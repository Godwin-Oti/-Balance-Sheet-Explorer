## Answering Potential Business Requirements

### 1. Total Assets and Growth

```python
# Extract total assets for both periods
total_assets_Jun_2024 = Consolidated_Balance_Sheets.loc[Consolidated_Balance_Sheets['Category'] == 'Total assets', 'Jun_30_2024'].values[0]
total_assets_Dec_2023 = Consolidated_Balance_Sheets.loc[Consolidated_Balance_Sheets['Category'] == 'Total assets', 'Dec_31_2023'].values[0]

# Calculate percentage growth
assets_growth = ((total_assets_Jun_2024 - total_assets_Dec_2023) / total_assets_Dec_2023) * 100

print(f"Total Assets Growth: {assets_growth:.2f}%")
```

```python
import matplotlib.pyplot as plt

# Example data
total_assets_data = pd.DataFrame({
    'Date': ['Dec 31, 2023', 'Jun 30, 2024'],
    'Total_Assets': [106618.0, 112832.0]
})

# Line Chart
plt.figure(figsize=(10, 6))
plt.plot(total_assets_data['Date'], total_assets_data['Total_Assets'], marker='o', linestyle='-', color='b')
plt.title('Total Assets and Growth')
plt.xlabel('Date')
plt.ylabel('Total Assets (USD Millions)')
plt.grid(True)
plt.show()
```

### 2. Current Assets vs Current Liabilities (Liquidity Position)

```python
''''The current ratio is a key liquidity metric that shows the company’s ability to pay its short-term obligations'''
```

```python
# Extract total current assets and total current liabilities
current_assets_Jun_2024 = Consolidated_Balance_Sheets.loc[Consolidated_Balance_Sheets['Category'] == 'Total current assets', 'Jun_30_2024'].values[0]
current_liabilities_Jun_2024 = Consolidated_Balance_Sheets.loc[Consolidated_Balance_Sheets['Category'] == 'Total current liabilities', 'Jun_30_2024'].values[0]

current_assets_Dec_2023 = Consolidated_Balance_Sheets.loc[Consolidated_Balance_Sheets['Category'] == 'Total current assets', 'Dec_31_2023'].values[0]
current_liabilities_Dec_2023 = Consolidated_Balance_Sheets.loc[Consolidated_Balance_Sheets['Category'] == 'Total current liabilities', 'Dec_31_2023'].values[0]

# Calculate the current ratio for both periods
current_ratio_Jun_2024 = current_assets_Jun_2024 / current_liabilities_Jun_2024
current_ratio_Dec_2023 = current_assets_Dec_2023 / current_liabilities_Dec_2023

print(f"Current Ratio (Jun 2024): {current_ratio_Jun_2024:.2f}")
print(f"Current Ratio (Dec 2023): {current_ratio_Dec_2023:.2f}")
```

```python
import matplotlib.pyplot as plt

# Example data
liquidity_data = pd.DataFrame({
    'Category': ['Current_ratio_Dec', 'Current_ratio_Jun'],
    'Amount': [current_ratio_Dec_2023, current_ratio_Jun_2024]  
})

# Bar Chart
plt.figure(figsize=(10, 6))
plt.bar(liquidity_data['Category'], liquidity_data['Amount'], color=['skyblue', 'salmon'])
plt.title('Current Assets vs Current Liabilities')
plt.xlabel('Category')
plt.ylabel('Amount (USD Millions)')
plt.show()
```

### 3. Debt Position and Financial Leverage

```python
'''Analyzing how the company’s debt position has changed between periods and its potential impact on financial risk.'''
```

```python
# Extract total debt for both periods
debt_Jun_2024 = Consolidated_Balance_Sheets.loc[Consolidated_Balance_Sheets['Category'] == 'Debt and finance leases, net of current portion', 'Jun_30_2024'].values[0]
debt_Dec_2023 = Consolidated_Balance_Sheets.loc[Consolidated_Balance_Sheets['Category'] == 'Debt and finance leases, net of current portion', 'Dec_31_2023'].values[0]

# Calculate change in debt
debt_change = debt_Jun_2024 - debt_Dec_2023
print(f"Change in Debt: {debt_change:.2f} million USD")
```

```python
import matplotlib.pyplot as plt

# Example data
debt_data = pd.DataFrame({
    'Debt_period': ['Dec_2023', 'June_2024'],
    'Amount': [debt_Dec_2023, debt_Jun_2024] 
})

# Bar Chart
plt.figure(figsize=(10, 6))
plt.bar(debt_data['Debt_period'], debt_data['Amount'], color=['lightcoral', 'lightblue'])
plt.title('Debt Position')
plt.xlabel('Debt Type')
plt.ylabel('Amount (USD Millions)')
plt.show()
```

### 4. Cash Flow Position


```python
'''''Analyzing the cash and cash equivalents to understand the liquidity situation.'''
```

```python
# Extract cash and cash equivalents
cash_Jun_2024 = Consolidated_Balance_Sheets.loc[Consolidated_Balance_Sheets['Category'] == 'Cash and cash equivalents', 'Jun_30_2024'].values[0]
cash_Dec_2023 = Consolidated_Balance_Sheets.loc[Consolidated_Balance_Sheets['Category'] == 'Cash and cash equivalents', 'Dec_31_2023'].values[0]

cash_change = cash_Jun_2024 - cash_Dec_2023
print(f"Change in Cash and Cash Equivalents: {cash_change:.2f} million USD")
```

```python
import matplotlib.pyplot as plt

# Example data
cash_flow_data = pd.DataFrame({
    'Category': ['Cash_flow_Dec_2023', 'Cash_flow_Jun_2024'],
    'Amount': [cash_Dec_2023, cash_Jun_2024]  
})

# Stacked Bar Chart
plt.figure(figsize=(10, 6))
plt.bar(cash_flow_data['Category'], cash_flow_data['Amount'], color=['green', 'red', 'blue'])
plt.title('Cash Flow Position')
plt.xlabel('Cash Flow Category')
plt.ylabel('Amount (USD Millions)')
plt.show()
```

### 5. Property, Plant, and Equipment (PP&E)

```python
# Extract Property, Plant, and Equipment data
ppe_Jun_2024 = Consolidated_Balance_Sheets.loc[Consolidated_Balance_Sheets['Category'] == 'Property, plant and equipment, net', 'Jun_30_2024'].values[0]
ppe_Dec_2023 = Consolidated_Balance_Sheets.loc[Consolidated_Balance_Sheets['Category'] == 'Property, plant and equipment, net', 'Dec_31_2023'].values[0]

ppe_change = ppe_Jun_2024 - ppe_Dec_2023
print(f"Change in Property, Plant, and Equipment: {ppe_change:.2f} million USD")
```
